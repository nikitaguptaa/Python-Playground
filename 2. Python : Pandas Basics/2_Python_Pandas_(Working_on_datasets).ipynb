**Building own datasets**

```python
#Create First dataframe

data={'rno':[101,102,103,104,105],'name':['Anu','Suresh','Kalpesh','Dhuresh','Kuresh'],'age':[23,24,25,26,27]}
student=pd.DataFrame(data)
student
```

```python
#Create second dataframe

data1={'rno':[102,103,104],'phy':[56,67,78],'chem':[23,34,45],'maths':[98,76,56]}
marks=pd.DataFrame(data1)
marks
```

```python
#To merge two or more dataframe use inbuilt function merge() in pandas
df1=pd.merge(student,marks,on='rno')
df1
```

```python
#Outer join
df2=pd.merge(student,marks,on='rno',how='outer')
df2
```

```python
df2.isnull()
```

```python
df2.isnull().sum()
```

```python
#Concat used - when two tables required to connect directly irrespective of unique id
df3=pd.concat([student,marks])
```

```python
df3
```

```python
df3=pd.concat([student,marks],axis=1)
df3
```
