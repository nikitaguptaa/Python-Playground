# Dataframe =  To hold data in tabular means  i.e.  to hold in rows and columns
# it is a two dimensional array , it is stored data in matrix form

```python
#create dataframe from list ; Key:Values ; Key will be the column name and values will be column data

student=[[101,'Seema',23,'Bhopal'],[102,'Anu',24,'Mumbai'],[103,'Madan',25,'Indore']]
data1=pd.DataFrame(student,columns=['Roll_no','Name','Age','City'])
```

```python
data1
```

```python
#create dataframe from dictionary

dic={'Roll_no':[101,102,103,104],'Name':['Seema','Anu','Rohit','Suresh'],'Age':[23,24,26,27],'City':['Bhopal','Mumbai','Mumbai','Pune']}
S=pd.DataFrame(dic)
```

```python
print(S)
```

```python
#Access data from database
#To retrieve Name of all student

S['Name']
```

```python
S.City
```

```python
#Select Multiple column
#To access Name and Age of all student
S[['Name','Age']]
```

```python
#To Select row
#use iloc[rowindex] -- Index Location
S.iloc[2]
```

```python
#Cell value -- Intersection of row and column
S.iloc[2,1]
```

```python
#in loc we can use column name
S.loc[2,'Name']
```

```python
#to select NAme ,age and city of one and two column then use slicing
S.iloc[1:3,0:4]
```

```python
#delete column use del keyword
del S['Age']
```

```python
S
```

```python
#to delete the row use inbuilt function drop()
S1=S.drop(3)
```
